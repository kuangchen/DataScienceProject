I'm a big movie fan. Every Friday night, I sit down and decide what movie I am going to watch with my wife. I don't study the screenplay, because I don't have time to read and it's also not easy to judge. Instead, I would check which actor/actress are going to be featured in a movie. So a few question arises very naturally,
* given the actors list, is it possible to predict if an upcoming movie is worthwhile to watch? 
* which actor/actress are box office poison? and who makes up the winning formula for a movie?
* to what extent do the directors also affect a movie's performance?
* is it possible to predict the box office or ratings directly?

There have been serious studies (see the famous [Netflix Prize](http://en.wikipedia.org/wiki/Netflix_Prize)) to answer these questions. However, it is still interesting to build an easy-to-interpret model that is fast and sufficiently accurate. 

In this tutorial, we are going to explore the Naive Bayes Model. We first gather all the actors/actresses in a list, and represent each movie by a binary actor feature vector $F = [0, 1, 0, 1, 1, \dots]$, $i^{th}$ position of which indicates $i^{th}$ actor in the list played (1) or didn't play (0) in the movie. Using Bayes theorem, we calculate the conditional probability of a movie being worth (denoted by $W$) to watch given its feature vector, 

<center>
$P(W |F)= \dfrac{P(F| W)P(W)}{P(F) } $
</center>
$P(W)$ and $P(F)$ can be calculated in a straightforward way, by counting the fraction of "worth" movies and fraction of movies with feature vector $F$ out of the total movie pool:
<center>
$P(W)= \dfrac{\mathrm{\sum_{movies\,labeled\,W}{1}}}{\mathrm{\sum_{movies}{1}}} $
</center>

<center>
$P(W)= \dfrac{\mathrm{\sum_{movies\,labeled\,F}{1}}}{\mathrm{\sum_{movies}{1}}} $
</center>

The conditional probability is difficult to calculate, but can be nevertheless approximated by the assumption that the individual features are independent of each other, i.e.

<center>
$P(F | W) = \displaystyle{\prod_k P(F_k| W)}$
</center>
where $P(F_k| W)$ are easily learned by the algorithm (by counting the fraction of worthy movies with $k^{th}$ actor present or not).

## Loading the MovieLens dataset
We use the database from the [Grouplens project](http://grouplens.org/datasets/hetrec-2011/). ## Cleaning and Building the dataset

First, we would like to have the $Y$ vector, which is a binary value to indicate whether a movie is worth watching. Fortunately, the Grouplens project has collected the crtitcs ratings on Rotten Tomato in a column named  <code>rtAllCriticsRating</code> in <code>movie_df</code> dataframe. Then whether a movie is worthwhile to watch can be simply defined by comparing its rating to the mean rating over all movies. 

However, there are a few issues that need to be solved here. First, there are invalid values (represented by <code>"\\N"</code>) in the <code>rtAllCriticsRating</code> column, which we replace with <code>nan</code>, and remove with <code>dropna</code>. 

```python
print("Notice movie_df had \\N among the unique rating values\n%s" % movie_df.rtAllCriticsRating.unique())
movie_df.rtAllCriticsRating.replace(to_replace="\\N", value=nan, inplace=True)
movie_df.dropna(inplace=True)
```

Second, if we plot the histogram of the ratings, we notice the ratings consists of a (right skewed) distribution centered around 6. We also see a strange big stump at zero rating, which is probably data input error. (For instance, [_Twelve Monkeys_](http://www.rottentomatoes.com/m/12_monkeys/) has rating equal to zero in the database, but it is actually 7.4 out of 10 on Rotten Tomato.) We remove them too. 

```python
movie_df.rtAllCriticsRating = movie_df.rtAllCriticsRating.astype(float)
fig, ax = subplots(1, 1)
ax.set(xlabel="Rotten Tomato Critics Rating", ylabel="Frequency")
movie_df.rtAllCriticsRating.hist(ax=ax, figsize=(6, 4), bins=arange(0, 10, 0.5), edgecolor='white')
movie_df = movie_df[movie_df.rtAllCriticsRating>0.1]
```

Finally, the ratings are systematically higher for old movies which are made before 1980s. This makes sense as Rotten Tomato doesn't archive reviews for all the movies, and old movies that are reviewed tend to be 'classics'. (This important observation is pointed out by [Harvard CS109 Homework](http://nbviewer.ipython.org/github/cs109/content/blob/master/HW3_solutions.ipynb)). We choose not to include them, because the actors in those movies are generally not active playing in movies made in the near future. 

```python
movie_df.plot(x="year", y='rtAllCriticsRating', style='o', ylim=(0, 10), mec='None', alpha=0.2, xlim=(1900, 2010))
movie_df = movie_df[movie_df.year >= 1980]
```

After we clean the dataset, the mean rating is found to be 5.76. At last, we add a binary column <code>worthWatching</code>, which is the $Y$ vector.

```python
summary = movie_df.rtAllCriticsRating.describe()
movie_df["worthWatching"] = movie_df.rtAllCriticsRating > summary["mean"]
print("Quick summary of critics rating\n======================\n%s" % summary)
```

Next we build the feature matrix $X$ (n_sampels, n_features) from the dataframe <code>actor_df</code>. However, we notice the majority of the actors/actresses appear in less than 5 movies. 

```python
fig, ax = subplots(1, 1)
ax.set(xlabel="Number of movies per actor", ylabel="Frequency", yscale='log')
movies_per_actor = actor_df.actorID.value_counts()
movies_per_actor.hist(ax=ax, figsize=(6, 4), bins=arange(-1, 80), bottom=1, edgecolor='white')
```

Compared to the prolific actors/actresses, these movie actors are unlikely to appear in any upcoming movie, and thus shouldn't affect the prediction much. In addition, excluding them also makes the process of maching learning faster. 

```python
threshold = 5
prolific_actor_list = movies_per_actor[movies_per_actor>threshold].index
actor_df = actor_df[actor_df.actorID.isin(prolific_actor_list)]
X = pd.crosstab(actor_df.movieID, actor_df.actorID)
```

At last, we take the overlap of the movies in $X$ and $Y$. Notice that the $X$ has over 8000 features in total.

```python
common = set(X.index) & set(movie_df.id)

X = X[X.index.isin(common)]
Y = movie_df.ix[movie_df.id.isin(common), "worthWatching"]
print("X.shape = %s, Y.shape = %s" % (X.shape, Y.shape))
```

## Training a Naive Bayes Model

We use the Bernoulli Naive Bayes model provided by <code>sklearn</code>, and using a 10-fold cross validation find a model accuracy of 67.5% without any tweaking of model parameters. Not too bad.

```python
import sklearn.naive_bayes
import sklearn.cross_validation
import sklearn.metrics
import sklearn.grid_search
import sklearn.tree
import sklearn.feature_selection
```

```python
clf = sklearn.naive_bayes.BernoulliNB(fit_prior=True, binarize=None)
scores = sklearn.cross_validation.cross_val_score(clf, X, Y, cv=10, scoring='accuracy')
print("Accuracy: %0.3f (+/- %0.3f)" % (scores.mean(), scores.std() * 2))
```

Before we consider tweaking the model, we see that the total number of features (actors) is high quite (over 8000), but are they all equally relevant? 

To answer this question, we select the features with the highest $N$ [$\chi^2$ scores](http://nlp.stanford.edu/IR-book/html/htmledition/feature-selectionchi2-feature-selection-1.html), and use them to train and cross-validate the Naive Bayes model to get an accuracy estimation $p(N)$. We plot $p(N)$ for a range of $N$ below,

```python
def NBScoreOnKBest(X, Y, k):    
    score = []
    kf = sklearn.cross_validation.KFold(n=X.shape[0], n_folds=5, shuffle=True, random_state=None)

    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
    
        kbest = sklearn.feature_selection.SelectKBest(sklearn.feature_selection.chi2, k).fit(X_train, Y_train)
        X_train_new = kbest.transform(X_train)
        X_test_new = kbest.transform(X_test)

        clf = sklearn.naive_bayes.BernoulliNB(fit_prior=True)
        clf.fit(X_train_new, Y_train)
        score.append(clf.score(X_test_new, Y_test))
    return array(score)
```

```python
k_list = list(range(400, 2900, 400))
s_list = [NBScoreOnKBest(X.as_matrix(), Y.as_matrix(), k) for k in k_list]
s_list = array(s_list)
fig, ax = subplots(1, 1)
ax.errorbar(k_list, s_list.mean(axis=1), yerr=s_list.std(axis=1))
ax.set(xlabel="num of important features to train NB model", ylabel="accuracy", ylim=(0.56, 0.71), xlim=(300, 2900), title="Prediction Accuracy")
ax.grid()
```

The figures shows we only need to use the most important 2000 features, while retaining essentially the same prediction accruacy. This agrees with our intuition: typically, you don't need to scan the whole actors list to guess if a movie's good. Although it is true that Naive Bayes is robust against these features, it is still beneficial to exclude them for faster computation speed.

( We can also try to use <code>sklearn.grid_search</code> to search for the optimal value of <code>alpha</code>, the smoothing parameter for the <code>BernoulliNB</code> classifer. However, it doesn't improve model accuracy. ) 

Now we can answer which actors would make a movie worth or not worth watching. We can construct an identity matrix of shape (n_actors, n_actors), where $i^{th}$ movie only features $i^{th}$ actor. The predicted probability of this feature matrix is thus equal to the correlation of the actor/actress with a movie's success. We sort the predicted probability to obtain the top and bottom 20 actors/actresses:

```python
kbest = sklearn.feature_selection.SelectKBest(sklearn.feature_selection.chi2, 2000).fit(X.as_matrix(), Y)
X_new = kbest.transform(X)
clf = sklearn.naive_bayes.BernoulliNB(fit_prior=True, alpha=0.1).fit(X_new, Y)
f = eye(X_new.shape[1])
p = clf.predict_proba(f)
rank = p[:, 1].argsort()[::-1]
indices = kbest.get_support(True)
print("Top 20 movie actors for worthy movies are:\n" + "\n".join(X.columns[indices[rank]][:20].tolist()))
print("\nTop 20 movie actors for unworthy movies are:\n" + "\n".join(X.columns[indices[rank]][-20:].tolist()))
```

## Directions for Improvements

Can we do better than 67% prediction accuracy? We will breifly discuss about a few ideas below. 

We can try to expand the feature space, such that in the enlarged feature space the worth and unworth movies may become better separated. One can assume that _major_ actores/actresses have a stronger correlation with a movie's success. To test this theory, one can simply expand $F$ to twice its current length, where the first (second) half denotes if an actor/actress has played in a major (minor) role. Unfortnuately, this separation has been proved not to increase the prediction efficiency. I have tested this hypothesis, and it doesn't improve the accuracy. 

One might also argue that directors play an equal or greater role in the sucess of a movie. To test this theory, we load the <code>movie_director.dat</code> dataframe, and simply combine it with the actor dataframe to form a bigger feature matrix $XX$.

```python
director_df = pd.read_csv("movie_directors.dat", delimiter="\t", encoding="ISO-8859-1")
director_df.head(2)
```

```python
director_df = director_df[director_df.movieID.isin(X.index)]
XX = pd.crosstab(director_df.movieID, director_df.directorID)
XX = X.join(XX, rsuffix="_")
clf2 = sklearn.naive_bayes.BernoulliNB(fit_prior=True)
scores = sklearn.cross_validation.cross_val_score(clf2, XX, Y, cv=5, scoring='accuracy')
print("Accuracy: %0.3f (+/- %0.3f)" % (scores.mean(), scores.std() * 2))
```

Same as previous section, we use Bernoulli Naive Bayes classifier, and are able to improve the accuracy to 69%. Of course, this is at the expense of increasing feature size (we need to keep track of top 4000 important features), thus causing considerably computation time to train the model.

We can also take a divide and conquer approach: instead of use one Bayesian classifier on the entire dataset, it is possible to split the original data into two smaller parts, which two different Bayesian classifiers are applied and work better (higher accuracy) on these two smaller problems. If this principle could work, then we can use it recursively: the smaller problem can be further divided into even smaller problem, and so on. If we do not gain any improvement of accuracy, we stop and use one Bayesian classifier. Eventually we are going to get a tree based on these successive split decisions. This is essentially a decision tree with Bayesian classifier accuracy as the split criterion, and it is invented by Kohavi and named *NBTree* (Naive Bayesian Tree). Unfortunately, no split prove to improve the accuracy. 

## Conclusion

We use Naive Bayesian classifier to achieve near 70% accuracy. A few modifications have been proposed but no significant accuracy improvement has been realized. The reason could be that the $W$ and $NW$ class are not well separated. If that's the case, then there seems to be no good way to solve it. The other reason could be the poor assumption of all features being independent. For instance, Robert De Niro and Al Pacino appear in [many films](http://www.imdb.com/list/ls000079722/) *together*. Let's assume their appearance are perfectly correlated, then Naive Bayesian would under estimate $P(\textrm{W|RDN, AP})$, because
<center>
$P(\textrm{W|RDN, AP}) = P(\textrm{W|RDN}) = P(\textrm{W|AP}) \gt P(\textrm{W|RDN}) \times P(\textrm{W|AP})$
</center>
If this is the case, then one needs to apply the domain knowledge, or apply more sophisicated model to learn the interdependencies.
