I'm a big movie fan. Every Friday night, I sit down and decide what movie I am going to watch with my wife. I don't study the screenplay, because I don't have time to read and it's also not easy to judge. Instead, I would check which actor/actress are going to be featured in a movie. So a few question arises very naturally,
* given the actors list, is it possible to predict if an upcoming movie is worthwhile to watch? 
* which actor/actress are box office poison? and who makes up the winning formula for a movie?
* to what extent do the directors also affect a movie's performance?
* is it possible to predict the box office or ratings directly?

There have been serious studies (see the famous [Netflix Prize](http://en.wikipedia.org/wiki/Netflix_Prize)) to answer these questions. However, it is still interesting to build an easy-to-interpret model that is fast and sufficiently accurate. 

In this tutorial, we are going to explore the Naive Bayes Model, which treat whether an actor/actress appears in movie as feature. This model attempt to assign a class label (worth or not worth watching) to $i^{th}$ movie with conditional probability
<center>
$\mathrm{Label(i)} = \mathrm{argmax}_j P(j |\mathrm{actors\, in\, i^{th}\ movie}) = \mathrm{argmax}_j P(\mathrm{actors\, in\, i^{th}\ movie} | j)P(j)$
</center>
Further this model assumes that given the class label, actors' contributions are _independent_ of each other. So the above equation can be further simplified into 

<center>
$\mathrm{Label(i)} = \mathrm{argmax}_j \prod_k P(\mathrm{k^{th}\, actor} | j)P(j)$
</center>

where the terms on the right hand side are easily learned by the algorithm. This tutorial is organized in the following structure:

## Loading the MovieLens dataset
We use the database from the [Grouplens project](http://grouplens.org/datasets/hetrec-2011/). The two relevant datbases are <code>movies.dat</code> and <code>movie_actors.dat</code>, which includes detailed information about movies from 1970 to present day, and the actors/actresses in those movies. We load them using <code>pandas</code> package.

```python
movie_df = pd.read_csv("movies.dat", delimiter="\t", encoding="ISO-8859-1")
movie_df.head(2)
```

```python
actor_df = pd.read_csv("movie_actors.dat", delimiter="\t", encoding="ISO-8859-1", usecols=[0, 1, 2])
actor_df.head(2)
```

## Cleaning and Building the dataset

First, we would like to have the $Y$ vector, which is a binary value to indicate whether a movie is worth watching. Fortunately, the Grouplens project has collected the crtitcs ratings on Rotten Tomato in a column named  <code>rtAllCriticsRating</code> in <code>movie_df</code> dataframe. Then whether a movie is worthwhile to watch can be simply defined by comparing its rating to the mean rating over all movies. 

However, there are a few issues that need to be solved here. First, there are invalid values (represented by <code>"\\N"</code>) in the <code>rtAllCriticsRating</code> column, which we replace with <code>nan</code>, and remove with <code>dropna</code>. 

```python
print("Notice movie_df had \\N among the unique rating values\n%s" % movie_df.rtAllCriticsRating.unique())
movie_df.rtAllCriticsRating.replace(to_replace="\\N", value=nan, inplace=True)
movie_df.dropna(inplace=True)
```

Second, if we plot the histogram of the ratings, we notice the ratings consists of a (right skewed) distribution centered around 6. We also see a strange big stump at zero rating, which is probably data input error. (For instance, [_Twelve Monkeys_](http://www.rottentomatoes.com/m/12_monkeys/) has rating equal to zero in the database, but it is actually 7.4 out of 10 on Rotten Tomato.) We remove them too. 

```python
movie_df.rtAllCriticsRating = movie_df.rtAllCriticsRating.astype(float)
fig, ax = subplots(1, 1)
ax.set(xlabel="Rotten Tomato Critics Rating", ylabel="Frequency")
movie_df.rtAllCriticsRating.hist(ax=ax, figsize=(6, 4), bins=arange(0, 10, 0.5), edgecolor='white')
movie_df = movie_df[movie_df.rtAllCriticsRating>0.1]
```

Finally, the ratings are systematically higher for old movies which are made before 1980s. This makes sense as Rotten Tomato doesn't archive reviews for all the movies, and old movies that are reviewed tend to be 'classics'. (This important observation is pointed out by [Harvard CS109 Homework](http://nbviewer.ipython.org/github/cs109/content/blob/master/HW3_solutions.ipynb)). We choose not to include them, because the actors in those movies are generally not active playing in movies made in the near future. 

```python
movie_df.plot(x="year", y='rtAllCriticsRating', style='o', ylim=(0, 10), mec='None', alpha=0.2)
movie_df = movie_df[movie_df.year >= 1980]
```

After we clean the dataset, the mean rating is found to be 5.76. At last, we add a binary column <code>worthWatching</code>, which is the $Y$ vector.

```python
summary = movie_df.rtAllCriticsRating.describe()
movie_df["worthWatching"] = movie_df.rtAllCriticsRating > summary["mean"]
print("Quick summary of critics rating\n======================\n%s" % summary)
```

Next we build the feature matrix $X$ (n_sampels, n_features) from the dataframe <code>actor_df</code>. However, we notice the majority of the actors/actresses appear in less than 5 movies. 

```python
fig, ax = subplots(1, 1)
ax.set(xlabel="Number of movies per actor", ylabel="Frequency", yscale='log')
movies_per_actor = actor_df.actorID.value_counts()
movies_per_actor.hist(ax=ax, figsize=(6, 4), bins=arange(-1, 80), bottom=1, edgecolor='white')
```

Compared to the prolific actors/actresses, these movie actors are unlikely to appear in any upcoming movie, and thus shouldn't affect the prediction much. In addition, excluding them also makes the process of maching learning faster. 

```python
threshold = 5
prolific_actor_list = movies_per_actor[movies_per_actor>threshold].index
actor_df = actor_df[actor_df.actorID.isin(prolific_actor_list)]
X = pd.crosstab(actor_df.movieID, actor_df.actorID)
```

At last, we take the overlap of the movies in $X$ and $Y$.

```python
common = set(X.index) & set(movie_df.id)

X = X[X.index.isin(common)]
Y = movie_df.ix[movie_df.id.isin(common), "worthWatching"]
print("X.shape = %s, Y.shape = %s" % (X.shape, Y.shape))
```

## Training a Naive Bayes Model

We use the Bernoulli Naive Bayes model provided by <code>sklearn</code>, and using a 3-fold cross validation find a model accuracy of 67% without any tweaking of model parameters. 

```python
import sklearn.naive_bayes
import sklearn.cross_validation
import sklearn.metrics
import sklearn.grid_search
import sklearn.tree
```

```python
clf = sklearn.naive_bayes.BernoulliNB(fit_prior=True)
scores = sklearn.cross_validation.cross_val_score(clf, X, Y, cv=5, scoring='accuracy')
print("Accuracy: %0.3f (+/- %0.3f)" % (scores.mean(), scores.std() * 2))
```

Not too bad. There is an <code>alpha</code> smoothing parameter for the <code>BernoulliNB</code> classifer. The grid search for the best value of <code>alpha</code> shows the accuracy can only be slighly improved. 

```python
parameters = {'alpha': arange(0, 2, 0.1)}
clf_ = sklearn.naive_bayes.BernoulliNB(fit_prior=True)
clf = sklearn.grid_search.GridSearchCV(clf_, parameters, n_jobs=3, cv=5)
clf.fit(X, Y)
print("best alpha = %3.2f, best score = %3.3f" % (clf.best_params_['alpha'], clf.best_score_))
```

The __R__eceiver __O__perating __C__haracteristic curve is shown below. The total area under the curve (auc) is found to be 0.92, also indicating the model gives good result (see References). 

```python
best_clf = clf.best_estimator_
pred_prob = best_clf.predict_proba(X)[:, 1]
fpr, tpr, threshold = sklearn.metrics.roc_curve(Y, pred_prob)
fig, ax = subplots(1, 1)
ax.plot(fpr, tpr)
ax.plot(fpr, fpr, "-")
ax.set(xlabel="False Positive Rate", ylabel="True Positive Rate")
print("Area under the curve = %3.2f" % sklearn.metrics.roc_auc_score(Y.data, pred_prob, average='macro', sample_weight=None))
```

Finally, we can find out which actors would make a movie worth or not worth watching. We can construct an identity matrix of shape (n_actors, n_actors), where $i^{th}$ movie only features $i^{th}$ actor. The predicted probability of this feature matrix is thus equal to the probability a specific actor/actress would make a movie worth or not worth watching. 

```python
f = eye(X.shape[1])
p = clf.predict_proba(f)
rank = p[:, 1].argsort()[::-1]
print("Top 20 movie actors for worthy movies are:\n")
print("\n".join(X.columns[rank][:20].tolist()))
print("\nTop 20 movie actors for unworthy movies are:\n")
print("\n".join(X.columns[rank][-20:].tolist()))
```

## Predicting based on directors?

One might argue that directors play an equal or greater role in the sucess of a movie. To test this theory, we load the <code>movie_director.dat</code> dataframe, and simply combine it with the actor dataframe to form a bigger feature matrix called $XX$. Same as previous section, we apply Bernoulli Naive Bayes classifier, and are able to  improve the accuracy to 69%. Of course, this is at the expense of increasing feature size, thus causing considerably computation time to train the model.

```python
director_df = pd.read_csv("movie_directors.dat", delimiter="\t", encoding="ISO-8859-1")
director_df.head(2)
```

```python
director_df = director_df[director_df.movieID.isin(X.index)]
XX = pd.crosstab(director_df.movieID, director_df.directorID)
XX = X.join(XX, rsuffix="_")
clf2 = sklearn.naive_bayes.BernoulliNB(fit_prior=True)
scores = sklearn.cross_validation.cross_val_score(clf2, XX, Y, cv=5, scoring='accuracy')
print("Accuracy: %0.3f (+/- %0.3f)" % (scores.mean(), scores.std() * 2))
```

## Further improvement of Naive Bayes Model
The Naive Bayes model performs poorly if the independence assumption between features is not met. For instance, Robert De Niro and Al Pacino appear in [many films](http://www.imdb.com/list/ls000079722/) together. Let's assume their appearance are perfectly correlated, then 

<center>
$P(\textrm{Worth | RDN, AP}) = P(\textrm{Worth | RDN}) = P(\textrm{Worth | AP}) \gt P(\textrm{Worth | RDN}) \times P(\textrm{Worth | AP})$
</center>

which violates the assumption. In this case, RDN and AP becomes essentially one feature, and the other redundant feature does not provide any additional information. This observation suggests that the identification and subsequent exclusion of redundant features could possibly improve the model accuracy. One possible method is use decision tree algorithm to pre-select important features (Ref 2), and then train the naive bayes model with those pre-selected features.
